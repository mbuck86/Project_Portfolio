# Introduction

This notebook is for creating the machine learning part of the analysis for my second Springboard Capstone project. Put simply, in this notebook I will use XGBoost to create a model that predicts MSRP of vehicles using the data that I have been working with throughout this project.
```python
# Importing packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# importing data
df = pd.read_csv('Cleaneddata.csv')
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
#Reorganizing the data so that MSRP is on the end of the df since it is the target variable
MS = df.MSRP
df1 = df.drop('MSRP', axis =1)
df1['MSRP'] = MS
```

# Data Preprocessing
Before I am able to do any modeling with the data, I will need to do a few things to it. Specifically I will need to change my categorical features into integers since XGBoost cannot use string categories. I will also standardize the numerical features.

```python
#defining categorical columns
cat_cols = ['make', 'model', 'year', 'fuel_type', 'num_cyl', 'trans_type', 'drive_type', 'doors','size', 'style', 'cat_1', 'cat_2']
```

```python
for col in cat_cols:
    df1[col] = pd.Categorical(df1[col])
```

```python
#ensuring data types are what they should be
df1.dtypes
```

```python
#create a loop that stores the categorical integers to a new df
int_df = pd.DataFrame()
```

```python
for col in cat_cols:
    col_name = col + '_code'
    int_df[col_name] = df1[col].cat.codes
```

```python
int_df.head()
```

Now that I have integer codes for each of the categorical variables, I'm going to standardize the numerical data.

```python
#import the standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
numerical_data = df1[['hp', 'highway_mpg', 'city_mpg', 'popularity', 'MSRP']]
```

```python
scaled_data = scaler.fit_transform(numerical_data)
```

```python
scaled_data = pd.DataFrame(scaled_data)
scaled_data.head()
```

```python
scaled_data.columns = ['hp', 'highway_mpg', 'city_mpg', 'popularity', 'MSRP']
scaled_data.head()
```

```python
new_dat = int_df.merge(scaled_data, right_index = True, left_index=True)
new_dat.head()
```

# Modifying the Data

Now that I have observed the out of the box feature importnaces for various models, I am going to take some time to modify the data that is being used, primarliy based on the advice of my mentor. 

1. Since it is known that Bugatti is a pretty extreme outlier, I am going to remove Bugatti from the data so thatthe importances of other features are not skewed by the Bugatti values.
2. I am going to limit the number of created dummy variables to 30 named types for both make and model, only keeping the most common of each. This could help improve model importance but also allow for an easier time evaluating feature importances. 
3. Impunting NaN values that I filled with zero in the earlier portions of this document.

"weight” is the number of times a feature appears in a tree

”gain” is the average gain of splits which use the feature

”cover” is the average coverage of splits which use the feature where coverage is defined as the number of samples affected by the split

```python
def variable_feat_train(n_feats, feat_list, Xtrain, ytrain, Xtest, ytest, model):
    """Takes a list of n numbers of features, the feature names, train test data, and a model, returns results for those n features"""
    results_list = []
    for n in n_feats:
        X_tr = Xtrain[feat_list[:n]]
        X_te = Xtest[feat_list[:n]]
        y_tr = ytrain
        y_te = ytest
        fit = model.fit(X_tr, y_tr)
        tr_pred = fit.predict(X_tr)
        te_pred = fit.predict(X_te)
        tr_rmse = np.sqrt(mean_squared_error(y_tr, tr_pred))
        te_rmse = np.sqrt(mean_squared_error(y_te, te_pred))
        results_list.append("N features: "+str(n)+" Training RMSE: " +str(tr_rmse) + " Testing RMSE: " + str(te_rmse) )
    return results_list
```
